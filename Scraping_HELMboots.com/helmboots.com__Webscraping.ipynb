```python
import pandas as pd
# ^^^ pyforest auto-imports - don't write above this line
import requests
import json
import pandas as pd
```

```python
url = 'https://helmboots.com/products.json?limit=250&page=1'
```

```python
r = requests.get(url)
```

```python
data = r.json()
```

```python
print(data)
```

```python
print(data['products'][0]['title'])
```

```python
for item in data['products']:
    print(item['title'])
```

```python
product_list = []


for item in data['products']:
    title = item['title']
    handle = item['handle']
    created = item['created_at']
    product_type = item['product_type']
    #print(title, handle, created, product_type)
    
    for image in item['images']:
        try:
            imagesrc = image['src']
        except: 
            imagesrc = 'None'
    for variant in item['variants']:
        price = variant['price']
        sku = variant['sku']
        available = variant['available']
        #print(price, sku, available)
        
        product = {
            'title': title,
            'handle':handle,
            'created':created,
            'product_type':product_type,
            'price':price,
            'sku':sku,
            'available':available,
            'image':imagesrc
        }
        
        
        product_list.append(product)
```

```python
print(product_list)
```

```python
df = pd.DataFrame(product_list)
```

```python
df.to_csv('helmboots.csv')
```
